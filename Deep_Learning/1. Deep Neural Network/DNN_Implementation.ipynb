```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
# Why XOR? Because it is a non-linearly seperable problem
# XOR is binary. Non-linearly -> at least one hidden layer.
# Neural network w/o hidden layer isn't able to solve non-linear problem.

# XOR problem training sample
training_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], "float32")
```

```python
# XOR problem target values accordingly
target_data = np.array([[0], [1], [1], [0]], "float32")
```

```python
# we can define the neural network layers in a sequential maner
model = Sequential()

# Don't have to define input layer explictly.
# It' going to be first hidden layer which means it's input layer
# first parameter is output dimension
model.add(Dense(16, input_dim = 2, activation='relu'))  # input layer has 2, output has 16
# it has 16 neurons in hidden layer.

model.add(Dense(16, input_dim = 16, activation='relu'))
model.add(Dense(16, input_dim = 16, activation='relu'))
model.add(Dense(16, input_dim = 16, activation='relu'))
model.add(Dense(16, input_dim = 16, activation='relu'))
model.add(Dense(16, input_dim = 16, activation='relu'))
model.add(Dense(16, input_dim = 16, activation='relu'))
model.add(Dense(16, input_dim = 16, activation='relu'))

# output layer.
model.add(Dense(1, activation='sigmoid'))
```

```python
# we can define the loss function MSE or negative log likelihood
model.compile(loss = 'mean_squared_error',
              optimizer = 'adam',  
              metrics=['binary_accuracy'])

# optimizer will find the right adjustments for the weights : SGD, Adgrad, ADAM... find out deeplearning4j.org/updater
# ADAM is..
# uses bothe first_ordered momentum mt and second-ordered g_t but they both decay over time.
# assign independent edge weight to each learning rate -> handle all learning-rate independently

# metric
# faroit.github.io/keras-docs/1.1.1/metrics/
# metric is a function that used to judge the performance of model.
# binary_accuracy : mean accuracy rate across all prediction for binary classification problems (XOR)
# categorical_accuracy : for multi class classification
# sparse_categorical
# top_k_categorical
```

```python
# epoch in an iteration over the entire dataset
# verbose 0 is silent 1, and 2 are showing results
model.fit(training_data, target_data, epochs=500, verbose=2)

# Make prediction with the trained nerual network
print(model.predict(training_data).round())
```
